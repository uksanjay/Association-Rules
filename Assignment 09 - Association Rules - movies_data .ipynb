```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

# Rule 1 -- Support = 0.1 , Confidence > 0.3 , Lift  > 1

```python
movies_df = pd.read_csv('F:\\PYTHON\\DATA SCIENCE\\Association rules\\ASSIGNMENT\\my_movies.csv')
```

```python
movies_df.shape
```

```python
movies_df
```

```python
movies_df = movies_df.iloc[: , 5:]
```

```python
movies_df
# Movies viewed by 10 users.
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_items = apriori(movies_df , min_support= 0.1 , use_colnames= True)
```

```python
frequent_items
# Filtering out combination having support > 0.1
```

```python
rule_1 = association_rules(frequent_items , metric= 'confidence' , min_threshold= .3)
```

```python
rule_1
```

```python
rule_1 = rule_1[rule_1['lift']>1]
# combinations with lift ratio > 1
```

```python
rule_1.sort_values('confidence' , ascending= False).reset_index(drop=True)
```

```python
plt.figure(figsize=(10,7))
sns.histplot([rule_1.confidence,rule_1.lift,rule_1.support],bins =50)
# Support ranges from 0 to .5 , lift ranges from 1 to 10 , confidence from .3 to 1
```

```python
sns.scatterplot(data=rule_1 , x='support' , y='confidence'  ,legend=False)
# Variation of support with confidence
```

# Rule 2 -- Support = 0.2 , Confidence > 0.4 , Lift  > 1

```python
frequent_items_new = apriori(movies_df , min_support= 0.2 , use_colnames=True)
```

```python
frequent_items_new
# Filtering items with min_support > 0.2
```

```python
rule_2 = association_rules(frequent_items_new , metric= 'confidence' , min_threshold= .4)
```

```python
rule_2_sorted = rule_2.sort_values('confidence',ascending=False).reset_index(drop=True)
```

```python
rule_2_sorted 
# List of most likely combinations of movies.
```

```python
pivot = rule_2_sorted.pivot('antecedents','consequents','confidence')
```

```python
pivot
```

```python
sns.heatmap(pivot ,annot=True)
```
